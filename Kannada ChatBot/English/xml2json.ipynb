```python
import xml.etree.ElementTree as ET
import json
```

```python
tree = ET.parse('crop.xml')
root = tree.getroot()
```

```python
[elem.tag for elem in root.iter()]
```

```python
data = {}
for elem in root.iter('Table'):
    temp = {}
    for ele in elem.iter():
        if ele.tag == 'Table':
            continue
        elif ele.tag == 'Min_x0020_Price':
            temp['Min_Price'] = ele.text.strip()
        elif ele.tag == 'Max_x0020_Price':
            temp['Max_Price'] = ele.text.strip()
        elif ele.tag == 'Modal_x0020_Price':
            temp['Modal_Price'] = ele.text.strip()
        else:
            temp[ele.tag] = ele.text.strip()

    print(temp)
    if temp['Commodity'] not in data:
        data[temp['Commodity'].lower()] = temp
    else:
        continue
```

```python
print(data)
```

```python
with open("crop_data_eng.json", "w") as write_file:
    json.dump(data, write_file, indent=4)
```
