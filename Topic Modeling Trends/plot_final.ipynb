# PLOT

```python
import matplotlib.pyplot as plt
```

```python
markers = ["o" , "^" , r'$\alpha$', "s", "P", "x", "*"]
```

## Plot 1

```python
cdc = [0, .262, .123, .116, .109, .112]
months = [ "Jan'20", "June'20", "Dec'20", "Jan'21", "June'21", "Dec'21" ]
```

```python
f = plt.figure()
f.set_figwidth(12)
f.set_figheight(6)

plt.scatter(months[0], cdc[0])
plt.plot(months[1:], cdc[1:], marker=markers[2])  # Plot the chart

plt.axvline(x=months[0], ymax=0.035, color='gray', linestyle='--')
plt.axvline(x=months[1], ymax=0.93, color='gray', linestyle='--')
plt.axvline(x=months[2], ymax=0.47, color='gray', linestyle='--')
plt.axvline(x=months[3], ymax=0.44, color='gray', linestyle='--')
plt.axvline(x=months[4], ymax=0.41, color='gray', linestyle='--')
plt.axvline(x=months[5], ymax=0.42, color='gray', linestyle='--')

plt.legend(["cdc"], loc=2, frameon=True)

plt.xlabel("Months")  # add X-axis label
plt.ylabel("Cosine Distance")  # add Y-axis label
plt.title("Base Term : 'Unofficial'")
plt.show()  # display
```
